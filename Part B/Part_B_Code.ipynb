# Modelling the data
#### To determine the drift rate (v) in the Drift Diffusion Model (DDM) based on user features like ExitRates, PageValues, etc., we can use a data-driven approach that links these features to the likelihood of purchase. Here's how to implement it:

Step 1: Model Purchase Probability with Logistic Regression

Train a logistic regression model to predict purchase probability using all relevant features. The model’s output (log-odds) will be used to compute the drift rate.

```python
# Cell 1: Imports and Data Preparation
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway, pearsonr
```

```python
# Assuming data is already loaded and preprocessed

# Cell 2: Feature Scaling and Model Training
# Features (already preprocessed)
data['rt'] = (data['Informational_Duration']+data['ProductRelated_Duration'])/60
X = data[["rt", "ExitRates", "PageValues", "Month", "VisitorType"]]

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train logistic regression
model = LogisticRegression(max_iter=1000)
model.fit(X_scaled, data["Revenue"])
```

```python
# Cell 3: Drift Rate Calculation
# Calculate purchase probabilities
p = model.predict_proba(X_scaled)[:, 1]

# Compute drift rates
sigma = 0.2
a = 1.0
p = np.clip(p, 1e-10, 1 - 1e-10)
v = (sigma**2 / (2 * a)) * np.log(p / (1 - p))

# Add drift rate to DataFrame immediately after calculation
data["v"] = v  # Critical fix: Add to DataFrame here
```

```python
# Cell 4: DDM Simulation
def simulate_ddm(v, sigma=0.2, a=1.0, z=0.5, T=10, dt=0.001):
    time_steps = np.arange(0, T, dt)
    x = [z]
    decision_time, decision_made = None, None

    for t in time_steps[1:]:
        noise = np.random.normal(0, sigma * np.sqrt(dt))
        new_x = x[-1] + v * dt + noise
        x.append(new_x)

        if new_x >= a:
            decision_time = t
            decision_made = 1
            break
        elif new_x <= 0:
            decision_time = t
            decision_made = 0
            break

    return decision_time, decision_made, x

# Simulate for all users
results = [simulate_ddm(vi) for vi in v]
data["Decision_Time"] = [res[0] for res in results]
data["Decision_Made"] = [res[1] for res in results]
data["Trajectory"] = [res[2] for res in results]
```

```python
# Cell 5: Product Engagement Categorization
# Create engagement categories
data["Product_Engagement"] = pd.qcut(
    data["ProductRelated_Duration"],
    q=[0, 0.25, 0.75, 1],
    labels=["Low", "Medium", "High"]
)

# Critical fix: Filter purchasers AFTER creating all columns
ddm_purchasers = data[data["Decision_Made"] == 1]
```

```python
# Cell 6: Basic Statistics
# Mean and SD calculations
print("DDM-Simulated Purchase Statistics:")
print(f"Mean Response Time: {ddm_purchasers['Decision_Time'].mean():.2f} minutes")
print(f"Response Time SD: {ddm_purchasers['Decision_Time'].std():.2f} minutes")

# Product engagement analysis
product_stats = ddm_purchasers.groupby("Product_Engagement").agg(
    Mean_Drift_Rate=("v", "mean"),
    Mean_Decision_Time=("Decision_Time", "mean"),
    Purchase_Rate=("Decision_Made", "mean")
).reset_index()

print("\nProduct Engagement Analysis:")
print(product_stats)
```

```python
# Cell 7: Visualizations
# Simulation vs Observed
plt.figure(figsize=(10, 6))
sns.kdeplot(ddm_purchasers["Decision_Time"]*60, label="DDM-Simulated", color="blue")
sns.kdeplot(data[data["Revenue"] == 1]["rt"], label="Observed", color="red")
plt.title("Response Time Distribution Comparison")
plt.xlabel("Time (seconds)")
plt.legend()
plt.show()

# Drift Rate Analysis
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(x="Product_Engagement", y="v", data=ddm_purchasers,
            order=["Low", "Medium", "High"])
plt.title("Drift Rates by Product Engagement")

plt.subplot(1, 2, 2)
sns.boxplot(x="Product_Engagement", y="Decision_Time", data=ddm_purchasers,
            order=["Low", "Medium", "High"])
plt.title("Decision Times by Product Engagement")
plt.tight_layout()
plt.show()
```

```python
 # Cell 8: Statistical Tests
# ANOVA Test
low = ddm_purchasers[ddm_purchasers["Product_Engagement"] == "Low"]["v"]
medium = ddm_purchasers[ddm_purchasers["Product_Engagement"] == "Medium"]["v"]
high = ddm_purchasers[ddm_purchasers["Product_Engagement"] == "High"]["v"]

f_stat, p_value = f_oneway(low, medium, high)
print(f"\nANOVA Results for Drift Rates:\nF-stat: {f_stat:.4f}, p-value: {p_value:.6f}")

# Correlation Analysis
corr, p = pearsonr(ddm_purchasers["v"], ddm_purchasers["Decision_Time"])
print(f"\nDrift Rate vs Decision Time Correlation:\nCorrelation: {corr:.4f}, p-value: {p:.6f}")
```

```python
# Cell 9: Individual Trajectories
# Plot random participants
random_indices = np.random.choice(len(data), 5, replace=False)

plt.figure(figsize=(15, 10))
for i, idx in enumerate(random_indices):
    row = data.iloc[idx]
    time_steps = np.arange(0, len(row["Trajectory"])) * 0.001

    plt.subplot(2, 3, i+1)
    plt.plot(time_steps, row["Trajectory"], label='Actual')
    plt.plot(time_steps, 0.5 + row["v"]*time_steps, '--', label='Expected')
    plt.axhline(1, color='r', linestyle=':')
    plt.axhline(0, color='b', linestyle=':')
    plt.title(f"User {idx+1}\nDrift Rate: {row['v']:.6f}")
    plt.xlabel("Time (s)")
    plt.ylabel("Decision Variable")
    plt.legend()

plt.tight_layout()
plt.show()
```
